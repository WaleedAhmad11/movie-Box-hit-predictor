```python
keep_cols = [
    "id", "title", "release_date", "original_language", "adult",
    "popularity", "vote_average", "vote_count", "genre_ids", "overview"
]

tmdb_clean = tmdb_df[keep_cols].copy()
tmdb_clean.head()
```

```python
import pandas as pd
import requests


TMDB_API_KEY = "8ec51db19c2537a6ef80f4816df58717"

def get_genre_mapping():
    url = f"https://api.themoviedb.org/3/genre/movie/list?api_key={TMDB_API_KEY}&language=en-US"

    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()["genres"]
        return {genre["id"]: genre["name"] for genre in data}
    else:
        print(f"Error fetching genres: {response.status_code}")
        return {}


genre_mapping = get_genre_mapping()

print(genre_mapping)
```

```python
import ast  # safely convert string lists to Python lists

def map_genres(genre_ids_str):
    try:
        ids = ast.literal_eval(genre_ids_str)  # convert string → list
        return [genre_mapping.get(i, "Unknown") for i in ids]
    except:
        return []

tmdb_clean["genres"] = tmdb_clean["genre_ids"].apply(map_genres)
```

```python
tmdb_clean['genres']
```

```python
tmdb_clean.drop(columns=['genre_ids'], inplace=True)
```

```python
tmdb_clean.info()
```

```python
# Make one column per genre (1 if movie has it, else 0)
for genre in sorted(set(g for lst in tmdb_clean['genres'] for g in lst if g)):
    tmdb_clean[genre] = tmdb_clean['genres'].apply(lambda x: int(genre in x))

```
